# Deployment

Du kannst die React-App auf deinem Raspberry Pi mit Nginx oder einem einfachen Node.js-Server hosten. Hier sind die Schritte:

### **1. React-App auf deinem lokalen Rechner builden**
Erstelle das Build-Verzeichnis:
```sh
conda activate react-ts-env
npm run build
```
Das generiert einen `dist/` Ordner mit den statischen Dateien.

```sh
scp -r dist pi@pi:/home/pi/Intranet
```

### **3. Nginx auf dem Raspberry Pi installieren**
Falls noch nicht installiert:
```sh
sudo apt update
sudo apt install nginx -y
```

### **5. Nginx aktivieren**
Erstelle einen Symlink zur Konfiguration:
```sh
sudo ln -s /etc/nginx/sites-available/react-app /etc/nginx/sites-enabled/
```
Starte Nginx neu:
```sh
sudo systemctl restart nginx
```
